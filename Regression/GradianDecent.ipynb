```python
def f(x):
    return x ** 3 + 4* x**2 - 3*x - 2

def df(x):
    return 3 * x **2 + 8*x - 3 
```

```python
iter = 1000
lr = 0.5
x = 6
list_x = []
```

```python
for i in range(iter):
    dx = -lr * df(x)
    x = x + dx
    list_x.append(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.arange(100), list_x[:100])
plt.show()
```

```python
def f(z):
    x, y = z
    return x ** 2 + y ** 2 + y - 1/2 * x * y

def df(z):
    x, y = z
    return np.array(2*x - 0.5*y, 2*y + 1 - 0.5*y)
```

```python
z = np.array((1, 2))
list_z = [z]
for i in range(iter):
    dz = -lr * df(z)
    z = z + dz
    list_z.append(z)
```

```python
X = np.array([[1, 1],
              [2, 1],
              [3, 1]])
Y = np.array([1, 2.1, 3]).reshape((3, 1))
w = np.array(([0],[0]))
lr = 0.01
```

```python
def loss_function(X, Y, w):
    return np.sum((X@w - Y)**2)

def gradient(X, Y, w):
    return 2 * X.T@(X@w-Y)
```

```python
w_list = [w]
for i in range(iter):
    dw = -lr * gradient(X, Y, w)
    w = w + dw
    w_list.append(w)
```
